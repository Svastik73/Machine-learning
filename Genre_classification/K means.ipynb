```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
#sources : coursera!!
df = pd.read_csv('dataset.csv')

null_= df.isnull().sum().sum()

if null_ == 0:
    print("No null values found in the GTZAN dataset.")
else:
    print(f"{null_} null values found in the GTZAN dataset.")
X = df.iloc[:, 1:-1].astype(float).values  # features as float
y = df.iloc[:, -1].values   # Labels

# PCA to reduce dimensionality
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
feature_names = df.columns[1:-1].values
for i in range(pca.n_components_):
    component = pca.components_[i]
    feature_idx = np.abs(component).argmax()
    feature_name = feature_names[feature_idx]
    print(f"Principal component {i+1} corresponds to feature '{feature_name}'")
colors = ['blue', 'green', 'red','yellow','pink','orange','purple']
    # number of clusters
K = 3
# Initialize the centroids randomly
centroids = X_pca[np.random.choice(X_pca.shape[0], K, replace=False)]
#source == coursera
for i in range(50):
    dist = np.sqrt(((X_pca - centroids[:, np.newaxis])**2).sum(axis=2))
    labels = np.argmin(dist, axis=0)
    for j in range(K):
        centroids[j] = np.mean(X_pca[labels == j], axis=0)
for k in range(K):
    cluster_samples = X_pca[labels == k]
    plt.scatter(cluster_samples[:, 0], cluster_samples[:, 1], color=colors[k])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='black')
plt.title('K-Means k==3')
plt.xlabel('Principal component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
